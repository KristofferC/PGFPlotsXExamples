## 5.8 Grouping plots

```julia
cs = [[(0,0), (1,1), (2,2)],
      [(0,2), (1,1), (2,0)],
      [(0,2), (1,1), (2,1)],
      [(0,2), (1,1), (1,0)]]

@pgf gp = GroupPlot({group_style = { group_size = "2 by 2",}, height = "6cm", width = "6cm"})

for coords in cs
    push!(gp, Plot(Coordinates(coords)))
end

gp
```
