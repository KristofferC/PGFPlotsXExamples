## 4.6 Three Dimensional Plot Types


```julia
import PGFPlotsX: Axis, Plot, Plot3, TikzPicture, TikzDocument,
                  Table, Expression, Coordinates, @pgf
using LaTeXStrings
```

### 4.6.12 Image (Matrix) Plots


```julia
x = repeat(0:2, outer = 3)
y = repeat(0:2, inner = 3)
meta = ["color=$c" for c in ["red", "blue", "yellow", "black", "brown", "magenta", "green", "red", "white"]]
c = Coordinates(hcat(x, y)'; metadata = meta)
@pgf Axis(Plot(c,
        {
            matrix_plot,
            mark = "*",
            nodes_near_coords = raw"\coordindex",
            "mesh/color input" = "explicit",
            "mesh/cols" = 3
        }), { enlargelimits = false, xtick = [0, 1, 2] }) # for some reason I need the xtick
```

```julia
c = Coordinates(hcat(x, y)'; metadata = collect(0:8))
@pgf Axis(Plot(c,
        {
            matrix_plot,
            mark = "*",
            nodes_near_coords = raw"\coordindex",
            point_meta = "explicit",
            "mesh/cols" = 3
        }), { enlargelimits = false, colorbar, xtick = [0, 1, 2] }) # for some reason I need the xtick
```

Using a `Table`:

```julia
t = Table(@pgf { meta = "C" }; x = x, y = y, C = collect(0:8))
@pgf Axis(Plot(t,
        {
            matrix_plot,
            mark = "*",
            nodes_near_coords = raw"\coordindex",
            point_meta = "explicit",
            "mesh/cols" = 3
        }), { enlargelimits = false, xtick = [0, 1, 2] }) # for some reason I need the xtick
```
