```julia
import PGFPlotsX: Axis, Plot, Plot3, 
                  PGFTable, PGFFunction, PGFCoordinates
using LaTeXStrings
```

```julia
Plot(PGFCoordinates([(0,0), (1,2), (2,3)]),
    "sharp plot")
```

```julia
Plot(PGFCoordinates([(0,0), (1,2), (2,3)]),
    "smooth")
```

```julia
c = PGFCoordinates([
(0,0.1), (0.1,0.15), (0.2,0.5), (0.3,0.62),
(0.4,0.56), (0.5,0.58), (0.6,0.65), (0.7,0.6),
(0.8,0.58), (0.9,0.55), (1,0.52)]);
```

```julia
Plot(c, "const plot")
```

```julia
Axis(Plot([c, "\\closedcycle"], 
        "const plot",
        "fill" => "blue",
        "draw" => "black";
        incremental = false),
    "ymin" => 0, 
    "ymax" => 1,
    "enlargelimits" => false)
```

```julia
Plot(c, "const plot mark right")
```

```julia
Plot(c, "const plot mark mid")
```

```julia
Axis([Plot(PGFFunction("4*x^2 - 5"),
        "jump mark left",
        "domain" => -5:0),
      Plot(PGFFunction("0.7*x^3 + 50"),
        "jump mark right",
        "domain" => -5:0)],
    "samples" => 8
)
```

```julia
Plot(c, "jump mark mid")
```
